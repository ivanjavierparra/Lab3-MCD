```python
df = pd.read_csv("../notebooks/entregable/datasets/periodo_x_producto_con_target.csv", sep=',', encoding='utf-8')
df
```

```python
# Convertimos la columna 'periodo' a tipo datetime
df['periodo_dt'] = pd.to_datetime(df['periodo'], format='%Y%m')

df_tsfresh = df[['product_id', 'periodo_dt', 'tn']].copy()

# Aseguramos el formato que espera tsfresh
df_tsfresh = df_tsfresh.rename(columns={
    'product_id': 'id',
    'periodo_dt': 'time',
    'tn': 'value'
})

# Extraemos características automáticamente
features = extract_features(
    df_tsfresh, 
    column_id='id', 
    column_sort='time',
    default_fc_parameters=EfficientFCParameters()
)

features
```

```python
# Imputar valores faltantes en las features (requerido antes de seleccionar)
impute(features)

# Suponiendo que tenés un vector y con la target por cada id
# selected_features = select_features(features, y)
```

```python
# Crear columna de tiempo tipo entero secuencial
df_ts = df.copy()
df_ts = df_ts[(df_ts['product_id']==20001) & (df_ts['periodo'] == 201701)]
df_ts['periodo_dt'] = pd.to_datetime(df_ts['periodo'], format='%Y%m')
# df_ts['time_idx'] = df_ts.groupby('product_id').cumcount()

# Renombrar columnas para usar con TSFRESH
df_ts = df_ts.rename(columns={'tn': 'value', 'product_id': 'id', 'periodo': 'time'})

# Extraer features (usa solo columnas: id, time, value)
features = extract_features(df_ts[['id', 'time', 'value']], 
                            column_id='id', 
                            column_sort='time',
                            disable_progressbar=False)

features
```

```python
features.to_json('features.json', orient='records', indent=2)
```
