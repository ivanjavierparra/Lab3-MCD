```python
import pandas as pd
import numpy as np
```

```python
archivo_lightgbm = pd.read_csv("./outputs/predicciones_exp_07_lgb_v2.csv", sep=',').rename(columns={'tn': 'pred_lgb'})
archivo_regresion_lineal = pd.read_csv("./outputs/predicciones_regresion_lineal_v1.csv", sep=',').rename(columns={'tn': 'pred_rl'})
archivo_autogluon = pd.read_csv("./outputs/prediccion_autogluon_2ventanas.csv", sep=',').rename(columns={'tn': 'pred_ag'})
```

```python
df = archivo_lightgbm.merge(archivo_regresion_lineal, on='product_id', how='left')
df = df.merge(archivo_autogluon, on='product_id', how='left')
df
```

```python
df['tn'] = 0.7 * df['pred_lgb'] + 0.2 * df['pred_rl'] + 0.1 * df['pred_ag']
df.drop(columns=['pred_lgb', 'pred_rl', 'pred_ag'], inplace=True)
df.to_csv("./outputs/predicciones_v8_ensemble_v1.csv", index=False, sep=',')
```
